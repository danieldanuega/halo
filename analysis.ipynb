```python
from matplotlib.image import imread
import matplotlib.pyplot as plt 
import os
```

```python
print("shape: {}".format(imread(dirr+filename).shape))
print("max value in image: {}".format(imread(dirr+filename).max()))

# Need to be normalize
```

```python
widths = []
heights = []

for image in os.listdir(dirr):
    img = imread(dirr+image)
    width, height, colors = img.shape
    widths.append(width)
    heights.append(height)
```

```python
import seaborn as sns

sns.jointplot(widths,heights)
```
